###Q1  What is the difference between a function and a method in Python ?

Function:

- A function is a block of code that performs a specific task and is defined using the def keyword.
- It can be called independently by its name.
- It may or may not take arguments and may or may not return a value.

Method:

- A method is a function that is associated with an object (an instance of a class).
- It is called using the dot notation on an object.
- It implicitly receives the object it's called on (usually as the self argument) and can operate on the object's data.

###Q4 What is the purpose of the `return` statement in a Python function?

The return statement in a Python function serves two main purposes:

1. Exiting the Function: When a return statement is encountered within a function, it immediately terminates the execution of the function and returns control to the point where the function was called.

2. Returning a Value: The return statement is used to send a value or values back from the function to the caller. This value can be of any data type (integer, string, list, object, etc.) or even another function.

###Q5 What are iterators in Python and how do they differ from iterables?

Iterables:

- Definition: An iterable is any Python object that can be looped over using a for loop. In simpler terms, it's something you can iterate through, one element at a time.
- Examples: Lists, tuples, strings, dictionaries, sets, files, and generators are all examples of iterables.
- Key Feature: Iterables have a special method called __iter__() that returns an iterator object. This iterator is then used to actually iterate over the elements.
Iterators:

- Definition: An iterator is an object that represents a stream of data. It provides a way to access the elements of an iterable sequentially without needing to know the underlying structure of the iterable.
- Key Methods: Iterators have two main methods:

__iter__(): Returns the iterator object itself. This is required for an object to be considered an iterator.

__next__(): Returns the next element in the stream. If there are no more elements, it raises a StopIteration exception.

###Q6  Explain the concept of generators in Python and how they are defined.

Generators are a special type of iterator in Python that provide a concise and memory-efficient way to create iterators. They are essentially functions that use the yield keyword to produce a sequence of values one at a time, instead of creating and storing the entire sequence in memory.

###Q7 What are the advantages of using generators over regular functions?

1. Memory Efficiency:

Generators: Produce values on demand (lazy evaluation), so they only store the current value and the state of the function. This is highly memory-efficient, especially for large sequences or infinite streams of data.
Regular Functions: Typically calculate and store the entire sequence of values in memory before returning it. This can lead to significant memory usage for large datasets.
2. Improved Performance:

Generators: Due to lazy evaluation, generators can avoid unnecessary computations, potentially leading to faster execution, especially if you don't need to process the entire sequence.
Regular Functions: Often calculate all values upfront, even if you only need a subset of them, which can be less efficient.
3. Representing Infinite Sequences:

Generators: Can represent infinite sequences easily, as they generate values one at a time, without needing to store the entire sequence in memory.
Regular Functions: Cannot represent infinite sequences directly, as they would require infinite memory to store the values.

4. Readability and Conciseness:

Generators: Often lead to more concise and readable code compared to using iterators explicitly, as the yield keyword simplifies the process of creating iterators.
Regular Functions: May require more complex logic to create and manage iterators.
5. Pipelining:

Generators: Can be easily chained together (pipelined) to create complex data processing workflows, where the output of one generator is fed as input to another.
Regular Functions: Pipelining with regular functions can be more cumbersome.

###Q9 Explain the purpose and usage of the `map()` function in Python.

```python
#1  Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
#the list.

def sumEven(lis):
  count=0

  for i in lis:
    if(i%2==0):
      count+=i
  return count

sumEven([1,2,3,4])
```

```python
#5 Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
#terms.

class FibonacciIterator:
    def __init__(self, num_terms):
        self.num_terms = num_terms
        self.current_term = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.current_term < self.num_terms:
            result = self.a
            self.a, self.b = self.b, self.a + self.b
            self.current_term += 1
            return result
        else:
            raise StopIteration

# Example usage
fibonacci_iterator = FibonacciIterator(10)
for number in fibonacci_iterator:
    print(number)

```

```python
#6 Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powerSquare(expo):
  for i in range(1,expo+1):
    yield 2**i

for x in powerSquare(5):
  print(x)

```

```python

#10 Create a Python program that uses `filter()` to remove all the vowels from a given string.
def remove_vowels(string):
    vowels = set('aeiouAEIOU')
    filtered_chars = filter(lambda char: char not in vowels, string)
    return ''.join(filtered_chars)


my_string = "hello"
result = remove_vowels(my_string)
print(result)

```
